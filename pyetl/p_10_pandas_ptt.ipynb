# Load files into DataFrame

```python
import os
import glob
import pandas as pd
```

```python
# os.listdir("/Users/uuboy.scy/side-project/tibame/cii101/pyetl/res_gossiping")
os.listdir("./res_gossiping")
```

```python
glob.glob(r"/Users/uuboy.scy/side-project/tibame/cii101/pyetl/res_gossiping/*.txt")
```

```python
data = []
error_file_path_list = []

for file_path in glob.glob("./res_gossiping/*.txt"):
    with open(file_path, "r", encoding="utf-8") as f:
        try:
            article_content = f.read()  # Read the whole article as string
        except:
            error_file_path_list.append(file_path)
        reply_str = article_content.split("---split---")[1]
        row_data = reply_str.split("\n")[1:-1]
        data.append(row_data)
        # print(row_data)
data
```

```python
columns = [
    "推",
    "噓",
    "分數",
    "作者",
    "標題",
    "時間",
]
```

```python
df = pd.DataFrame(
    data=data,
    columns=columns,
)

df
```

```python
"""
"時間: Sat Nov 23 21:23:39 2024"
    ->  Sat Nov 23 21:23:39 2024
    ->  Timestamp('2024-11-23 21:23:39')
"""
df["時間"].apply(lambda x: x.replace("時間: ", "")).apply(pd.to_datetime)

df["operation_datetime"] = (
    df["時間"]
        .apply(lambda x: x.replace("時間: ", ""))
        .apply(pd.to_datetime)
)

df
```
